```python
import pandas as pd
import matplotlib.pyplot as plt

tech_support_dataset = pd.read_csv("tech_support_dataset.csv")
tech_support_dataset.head()
```

```python
# Convert the Resolution Time data into integer

def extract_minutes(time_str):
    return int(time_str.split()[0])

# Apply the function to the 'Resolution_Time' column and convert to integer
tech_support_dataset['Resolution_Time'] = tech_support_dataset['Resolution_Time'].apply(extract_minutes).astype(int)
print(tech_support_dataset.head())
```

```python
tech_support_dataset.info()

checkUnique = tech_support_dataset.nunique()
print(checkUnique)
```

```python
# Check what are the 7 customer issues or concerns for calling

customerIssueSpecifics = tech_support_dataset[['Customer_Issue']].drop_duplicates()
print(customerIssueSpecifics)
```

```python
# Check what are the 7 tech responses for the customer's issue

techResponseSpecifics = tech_support_dataset[['Tech_Response']].drop_duplicates()
print(techResponseSpecifics)
```

```python
# Check what are the 5 categories in the Issue Category Column

customerIssueCategory = tech_support_dataset[['Issue_Category']].drop_duplicates()
print(customerIssueCategory)
```

```python
# Check what are the 4 categories in the Issue Status Column

customerIssueStatus = tech_support_dataset[['Issue_Status']].drop_duplicates()
print(customerIssueStatus)
```

```python
issue_category_counts = tech_support_dataset['Issue_Category'].value_counts()

plt.figure(figsize=(10, 6))  # Adjust figure size for better visualization
plt.bar(issue_category_counts.index, issue_category_counts.values)
plt.title('Figure 1. Count of Issues by Category')
plt.xlabel('Issue Category')
plt.ylabel('Count of Issues')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
issue_category_counts = tech_support_dataset['Issue_Category'].value_counts()

# Define a list of contrasting colors (adjust as needed)
colors = ['royalblue', 'gold', 'limegreen', 'firebrick', 'blueviolet']  # You can add more colors for more categories

# Create the bar plot with contrasting colors
plt.figure(figsize=(10, 8))  # Adjust figure size for better visualization
plt.bar(issue_category_counts.index, issue_category_counts.values, color=colors[:len(issue_category_counts)])  # Slice colors to match data length

plt.title('Figure 1. Count of Issues by Category')
plt.xlabel('Issue Category')
plt.ylabel('Count of Issues')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
issue_category_counts = tech_support_dataset['Issue_Category'].value_counts()
colors = ['royalblue', 'gold', 'limegreen', 'firebrick', 'blueviolet']

plt.figure(figsize=(12, 9))
bars = plt.bar(issue_category_counts.index, issue_category_counts.values, color=colors[:len(issue_category_counts)])

for bar, count in zip(bars, issue_category_counts):
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 5, int(count), ha='center', va='bottom', fontsize=10)

plt.title('Figure 1. Count of Issues by Category')
plt.xlabel('Issue Category')
plt.ylabel('Count of Issues')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
average_resolution_time = tech_support_dataset.groupby('Issue_Category')['Resolution_Time'].mean()

color1 = ['blueviolet', 'royalblue', 'limegreen', 'gold', 'firebrick']

plt.figure(figsize=(10, 8))

bars = plt.bar(average_resolution_time.index, average_resolution_time.values, color=color1[:len(issue_category_counts)])
for bar, yval in zip(bars, average_resolution_time):
    y_pos = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, y_pos + 0.1, round(yval, 2), ha='center', va='bottom')

plt.title('Figure 1.1 Average Resolution Time by Issue Category')
plt.xlabel('Issue Category')
plt.ylabel('Average Resolution Time (minutes)')
plt.xticks(rotation=60, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

color1 = ['blueviolet', 'royalblue', 'limegreen', 'gold', 'firebrick']  # You can add more colors for more categories


# Group by 'Issue_Category' and calculate the mean 'Resolution_Time'
average_resolution_time = tech_support_dataset.groupby('Issue_Category')['Resolution_Time'].mean()

# Create the bar chart
plt.figure(figsize=(10, 8))
plt.bar(average_resolution_time.index, average_resolution_time.values, color=color1[:len(issue_category_counts)])
plt.title('Figure 1.1 Average Resolution Time by Issue Category')
plt.xlabel('Issue Category')
plt.ylabel('Average Resolution Time (minutes)')
plt.xticks(rotation=60, ha='right')
plt.tight_layout()
plt.show()
```

```python
status_order = ['Pending', 'Resolved', 'Resolved after follow-up', 'Escalated']

issue_status_counts = tech_support_dataset.groupby(['Issue_Category', 'Issue_Status']).size().unstack(fill_value=0)

issue_status_counts = issue_status_counts[status_order]

issue_status_counts.plot(kind='bar', figsize=(16, 14))
plt.title('Figure 1.2 Issue Status by Category')
plt.xlabel('Issue Category')
plt.ylabel('Number of Issues')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.legend(title='Issue Status')
plt.show()
```

```python
plt.figure(figsize=(16, 10))

colors = ['skyblue', 'lightsalmon', 'lightgreen', 'cornflowerblue', 'plum', 'seagreen', 'aqua']

for i, response in enumerate(tech_support_dataset['Tech_Response'].unique()):
    subset = tech_support_dataset[tech_support_dataset['Tech_Response'] == response]
    plt.boxplot(subset['Resolution_Time'], positions=[i], patch_artist=True, boxprops=dict(facecolor=colors[i % len(colors)]))

plt.title('Figure 2. Resolution Time by Technical Response')
plt.xlabel('Technical Response')
plt.ylabel('Resolution Time (minutes)')
plt.xticks(range(len(tech_support_dataset['Tech_Response'].unique())), tech_support_dataset['Tech_Response'].unique(), rotation=55, ha='right')
plt.tight_layout()
plt.show()
```

```python
software_pending_df = tech_support_dataset[(tech_support_dataset['Issue_Category'] == 'Hardware') &
                                          (tech_support_dataset['Issue_Status'] == 'Escalated')]

# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(software_pending_df.index, software_pending_df['Resolution_Time'], c=software_pending_df['Resolution_Time'], cmap='viridis') #Color by resolution time

plt.title('Distribution of Resolution Time for Software Issues (Pending Status)')
plt.xlabel('Issue Index')
plt.ylabel('Resolution Time (minutes)')
plt.colorbar(label='Resolution Time')  # Add a colorbar to indicate resolution time values
plt.show()
```

```python
resolved_issues = tech_support_dataset[tech_support_dataset['Issue_Status'] == 'Resolved']
resolved_followup_issues = tech_support_dataset[tech_support_dataset['Issue_Status'] == 'Resolved after follow-up']

resolved_times = resolved_issues.groupby('Tech_Response')['Resolution_Time'].mean()
resolved_followup_times = resolved_followup_issues.groupby('Tech_Response')['Resolution_Time'].mean()

plt.figure(figsize=(15, 8))
plt.plot(resolved_times.index, resolved_times.values, marker='o', linestyle='-', label='Resolved')
plt.plot(resolved_followup_times.index, resolved_followup_times.values, marker='s', linestyle='--', label='Resolved after follow-up')
plt.title('Figure 2.1 Distribution of Resolution Times by Tech Response')
plt.xlabel('Tech Response')
plt.ylabel('Average Resolution Time (minutes)')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```
