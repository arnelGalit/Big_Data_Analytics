### Question 1: Is there a relationship between Gross Domestic Product per capita and Prevalence of thinness among adolescents aged 10-19

```python
# Is there a relationship between Gross Domestic Product per capita and Prevalence of thinness among adolescents aged 10-19

x = life_expect_df['GDP_per_capita']
y = life_expect_df['Thinness_ten_nineteen_years']
plt.scatter(y,x)
plt.ylabel('Gross Domestic Product per capita')
plt.xlabel('Prevalence of thinness among adolescents aged 10-19')
plt.title('Scatter Plot of Gross Domestic Product per capita and Prevalence of thinness among adolescents aged 10-19')
plt.show()

# Observations: The plot shows that there's no relationship between  Gross Domestic Product per capita and Prevalence
# of thinness among adolescents aged 10-19. The plots are distributed mostly at the lower right of the graph.
```

### Question 2: What are the frequencies of immunization rates among the regions?

```python
# What is the average of immunization rates for illnesses among the regions?

freq_immunization_region = life_expect_df.groupby('Region')[['Hepatitis_B', 'Measles', 'Polio', 'Diphtheria']].mean()
freq_immunization_region.plot(kind='bar')
plt.xlabel('Region')
plt.ylabel('Frequency of immunization rates')
plt.title('Frequency of immunization rates among the regions')
plt.show()

# Observations: As the bar plot shows, the European Union has the highest immunization rate whereas
# Africa has the lowest immunization rate. Polio and Diphtheria have the highest immunization rate
# among the 4 illnesses.
```
